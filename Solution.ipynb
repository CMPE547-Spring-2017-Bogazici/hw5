```octave
nu = 0.4;beta=250*0.4; nu_beta = nu/beta;
P = 1.2; x0 = 1;x1 = -6;
T = 100; % number of iterations
R_lin = linspace(10^-4,10^-1,800);
R_lin = [R_lin';linspace(10^-1,10^6,300)'];
A_lin = linspace(-4,4,100);
[R_x, R_a] = meshgrid(R_lin', A_lin);
Z = [];
i = 0;
for r=R_lin'
    l = 0;
    for a =A_lin
        e1=exp(-(x1^2)/(2*r)+x0*x1*a/r-a^2*x0^2/(2*r)-log(2*pi*r)/2);
        e2=exp(-(a^2)/(2*P)-log(2*pi*P)/2);
        e3=exp(-(nu+1)*log(r)-nu/(beta*r)-log(gamma(nu))+nu*log(nu/beta));
        l = [l;e1*e2*e3];
    end
    if i == 0
        Z = l';
        i = 1;
    else
        Z = vertcat(Z,l');
    end
end

contour(A_lin, R_lin', Z)
xlabel('A')
ylabel('R')
set(gca, 'YScale', 'log')
title('Contour of full joint probability')
```

```octave
#EM for R
nu = 0.4; beta = 250*0.4; P = 1.2; x_0 = 1; x_1 = -6;
T = 100; E_A = -6; E_A2 = E_A^2; E_invR = 0.1;

for i = 2:T
    Sigma = 1/(1/P + x_0^2*E_invR);
    mu    = Sigma*x_0*x_1*E_invR;
    E_A  = mu;
    E_A2 = mu^2 + Sigma;
    
    a = nu + 0.5;
    b = 0.5*(x_1^2 - 2*x_1*x_0*E_A + x_0.^2*E_A2) + nu/beta;
    
    E_invR = (a)/b;
    E_R = b/(a+1);
end
E_invR
E_R
E_A
```
